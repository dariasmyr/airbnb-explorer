```python
import matplotlib.pyplot as plt

# Входные данные (время и стоимость посещения)
time = [8, 10, 12, 14, 16, 18]
cost = [1000, 1500, 2000, 2500, 2000, 1500]

# Создаем график
plt.plot(time, cost)

# Настраиваем график
plt.title("Зависимость стоимости посещения СПА салона от времени")
plt.xlabel("Время посещения (в часах)")
plt.ylabel("Стоимость посещения (в рублях)")

# Показываем график
plt.show()
```
